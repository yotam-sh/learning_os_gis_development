```python
import pandas as pd
import numpy as np
```

```python
exam_data = {
    'name': ['Anastasia', 'Dima', 'Katherine', 'James', 'Emily', 'Michael', 'Matthew', 'Laura', 'Kevin', 'Jonas'],
    'score': [12.5, 9, 16.5, np.nan, 9, 20, 14.5, np.nan, 8, 19],
    'attempts': [1, 3, 2, 3, 2, 3, 1, 1, 2, 1],
    'qualify': ['yes', 'no', 'yes', 'no', 'no', 'yes', 'yes', 'no', 'no', 'yes']
    }
labels = ['a', 'b', 'c', 'd', 'e', 'f', 'g', 'h', 'i', 'j']

df = pd.DataFrame(data=exam_data, index=labels)
df
```

##### 25. Write a Pandas program to change the order of a DataFrame columns.

```python
df_reorder = df.copy()
cols = df.columns.tolist()
cols.sort()
df_reorder = df_reorder.reindex(columns=cols)
df_reorder
```

##### 26. Write a Pandas program to add one row in an existing DataFrame.

```python
df_add = pd.concat([df, pd.Series({'name': 'Yotam', 'score': 15, 'attempts': 1, 'qualify': 'yes'}).to_frame().T], ignore_index=True)
df_add
```

##### 28. Write a Pandas program to count city wise number of people from a given of data set (city, name of the person)

```python
df_2 = pd.DataFrame({
    'name': ['Anastasia', 'Dima', 'Katherine', 'James', 'Emily', 'Michael', 'Matthew', 'Laura', 'Kevin', 'Jonas'],
    'city': ['California', 'Los Angeles', 'California', 'California', 'California', 'Los Angeles', 'Los Angeles', 'Georgia', 'Georgia', 'Los Angeles']}
    )
print(df_2.value_counts(subset=['city']))
```

##### 29. Write a Pandas program to delete DataFrame row(s) based on given column value.

```python
df_drop = df.copy().drop([i for i, j in df.iterrows() if j[1] > 10])
df_drop
```

##### 31. Write a Pandas program to select a row of series/dataframe by given integer index.

```python
# using .iloc[[]] to select a number of rows --> j[1][1] extract from j the data of each row and then the 'score' value from the data.
df_sel = df.copy().iloc[[i for i, j in enumerate(df.iterrows()) if j[1][1] > 10]]
df_sel
```

##### 32. Write a Pandas program to replace all the NaN values with Zero's in a column of a dataframe.

```python
df_nan = df.copy()
df_nan['score'] = df_nan['score'].replace({np.nan: np.zeros(1)})
df_nan
```
